```python
import glob
glob.glob('train/*.csv')#"*"はワイルドカード
```

```python
file_list=glob.glob('train/*.csv')
original_df=pd.DataFrame()

for file in file_list:
    df_temp=pd.read_csv(file)
    original_df=pd.concat([original_df,df_temp])
```

```python
original_df
```

```python
#全部欠損値の行削除
original_df=original_df.dropna(axis=1,how='all')
original_df.info()
```

```python
#欠損値の数確認
original_df.isnull().sum()
```

```python
#統合したものをエクセル保存
#df.to_excel('data_merged.xlsx',index=None)
```

```python
df=original_df.copy()
# 面積を数値型
df['面積（㎡）'] = df['面積（㎡）'].replace('2000㎡以上', 2000)
df['面積（㎡）']= pd.to_numeric(df['面積（㎡）'], errors='coerce')
#距離の数値変換と欠損値や？の削除、
df['最寄駅：距離（分）']= pd.to_numeric(df['最寄駅：距離（分）'], errors='coerce')
df.dropna(subset=['最寄駅：距離（分）'], inplace=True)
#建蔽率、容積率の欠損値削除、
# '建ぺい率（％）'または'容積率（％）'のどちらかに欠損値がある行を削除
df.dropna(subset=['建ぺい率（％）', '容積率（％）'], inplace=True)
```

```python
# --- 3. 処理後のデータを保存 ---
# 保存先のフォルダとファイル名を指定
processed_dir = '../data/processed/'
processed_file = 'cleaned_data.parquet'
processed_path = os.path.join(processed_dir, processed_file)

# processedフォルダがなければ作成する
os.makedirs(processed_dir, exist_ok=True)

# Parquet形式で保存（次の工程に引き継ぐ）
df.to_parquet(processed_path, index=False)

print(f"\nクレンジング後のデータを {processed_path} に保存しました。")
```
